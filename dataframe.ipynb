```python
# 지역별로 ['졸업자', '진학자', '일반고', '특성화고', '특목고', '자율고', '기타', '취업자', '무직자및미상']만 고르기.

a = '서울 부산 대구 인천 광주 대전 울산 세종 경기 강원 충북 충남 전북 전남 경북 경남 제주'
target = a.split(' ')
df2 = pd.DataFrame()

for j in target:
    df = pd.read_excel(f"./datasets/19 {j}.xlsx")
    df1 = pd.DataFrame(df.loc[8])
    df1.drop(['Unnamed: 1','Unnamed: 4'], axis=0, inplace=True)
    df1.rename(index={'Middle School': "Unnamed: 11"}, inplace=True)
    df1.reset_index(inplace = True)
    df1.drop(['index'], axis=1, inplace=True)

    for i in df1.index:
        if i % 2 == 1:
            df1.loc[i]
        else:
            df1.drop([i],axis=0,inplace=True)

    df1.index = ['졸업자', '진학자', '일반고', '특성화고', '특목고', '자율고', '기타', '취업자', '무직자및미상']

    df1 = df1.T

#     df1.rename(index = {8:f'{j}'})
    
    df2 = pd.concat([df2, df1])

df2.index = target

df19 = df2.copy()
```

### 진학 비율, 사교육비, 음주율, 우울감경험률, 흡연율, 비만율을 Indicator 컬럼으로 넣기

```python
df = pd.read_csv('./datasets/18_22비만율추가.csv')
```

```python
df
```

```python
df2 = pd.DataFrame()
```

```python
columns = ['특성화고진학률','특목고진학률','자율고진학율','특+자 진학률','사교육비','음주율','우울감경험률','흡연율','비만율']
values = ['특성화고','특목고','자율고']
i = 0

for i in df.index:
    for j in range(0, 9):
        k = i*9 + j
        df2.loc[k,'지역'] = df.loc[i,'지역']
        df2.loc[k,'Indicator Name'] = f'{columns[j]}'
        df2.loc[k,'year'] = str(df.loc[i,'year'])
        if j <= 2:
            df2.loc[k,'value'] = df.loc[i,f'{values[j]}']/df.loc[i,'졸업자']
        elif j == 3:
            df2.loc[k,'value'] = (df.loc[i,f'{values[j-2]}'] + df.loc[i,f'{values[j-1]}'])/df.loc[i,'졸업자']
        elif j == 4:
            df2.loc[k,'value'] = df.loc[i,'사교육비']
        elif j == 5:
            df2.loc[k,'value'] = df.loc[i,'음주율']
        elif j == 6:
            df2.loc[k,'value'] = df.loc[i,'우울감경험률']
        elif j == 7:
            df2.loc[k,'value'] = df.loc[i,'흡연율']
        elif j == 8:
            df2.loc[k,'value'] = df.loc[i,'비만율']
    i += 1
```

```python
df2 # 결과 확인
```

```python
df2['year'] = df2['year'].apply(lambda x: int(x)) # year 컬럼 값을 정수로.
```

```python
df2.to_csv('./datasets/Indicator18_22.csv', index = False) # 'Indicator18_22.csv'로 저장.
```
